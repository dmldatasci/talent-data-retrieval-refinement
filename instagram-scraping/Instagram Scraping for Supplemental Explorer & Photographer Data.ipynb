```python
# load dependencies
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

import requests
from bs4 import BeautifulSoup

import re
import pandas as pd
```

```python
# load talent data
talent_df = pd.read_csv('datasets/nonmale_wIG_nonchannels.csv')
# iterate through rows of talent
for index, talent in talent_df.iterrows():
    
    # find instances where talent only entered their IG handle, leading with a '@'
    if '@' in talent['Instagram']:
        talent_df.iloc[index]['Instagram'] = "https://www.instagram.com/" + talent['Instagram'].split('@')[1].split()[0]
    
    # find instances where talent did not enter a legitimate url, and did not lead their IG handle with a '@'
    elif 'instagram.com' not in talent['Instagram']:
        talent_df.iloc[index]['Instagram'] = "https://www.instagram.com/" + talent['Instagram'].split('/')[len(talent['Instagram'].split('/')) - 1].split()[0]

# establish acceptable residential locations
st_abbr = ['CT', 'NJ', 'VT', 'DE', 'NY', 'VA', 'GA', 'NC', 'WV', 'ME', 
           'OH', 'FL', 'MD', 'PA', 'IN', 'MA', 'RI', 'MI', 'NH', 'SC']

states = ['Connecticut', 'New Jersey', 'Vermont', 'Delaware', 'New York', 
          'Virginia', 'Georgia', 'North Carolina', 'West Virginia', 'Maine',
          'Ohio', 'Florida', 'Maryland', 'Pennsylvania', 'Indiana',
          'Massachusetts', 'Rhode Island', 'Michigan', 'New Hampshire', 'South Carolina']
```

```python
# scrapy scraper
def RetrieveFollowerCountBS(url):
    html = requests.get(url)
    soup = BeautifulSoup(html.text, 'html.parser').prettify().split(' Followers')[0].split('"')
    fin_soup = re.sub(',', '', soup[len(soup) - 1])
    
    print("fin_soup is ", fin_soup)
    
    if 'M' in fin_soup:
        approx_count = int(fin_soup.split('M')[0]) * 1000000
        
    elif 'K' in fin_soup:
        approx_count = int(fin_soup.split('K')[0]) * 1000
        
    else:
        try:
            approx_count = int(fin_soup)
        except: 
            approx_count = False
    
    return approx_count


ig_followers = [RetrieveFollowerCountBS(ig_link) for ig_link in talent_df['Instagram']]
```
